## load the JSON file that you created in the previous exercise into a dataframe

```python
import pandas as pd

# 1. –ó–∞–≥—Ä—É–∑–∫–∞ JSON-—Ñ–∞–π–ª–∞
df = pd.read_json("../ex02/auto.json")

# 2. –£—Å—Ç–∞–Ω–æ–≤–∫–∞ CarNumber –∫–∞–∫ –∏–Ω–¥–µ–∫—Å–∞
df.set_index("CarNumber", inplace=True)
```

## make the following selects

```python
# 1 –®—Ç—Ä–∞—Ñ—ã –±–æ–ª—å—à–µ 2100
fines_gt_2100 = df[df['Fines'] > 2100]
print("üöó –°—Ç—Ä–æ–∫–∏ —Å Fines > 2100:")
print(fines_gt_2100, "\n")

# 2 –®—Ç—Ä–∞—Ñ—ã > 2100 –∏ Refund == 2
fines_gt_2100_refund_2 = df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
print("üöó –°—Ç—Ä–æ–∫–∏ —Å Fines > 2100 –∏ Refund == 2:")
print(fines_gt_2100_refund_2, "\n")

# 3 –ú–æ–¥–µ–ª–∏ –∏–∑ —Å–ø–∏—Å–∫–∞
models_list = ['Focus', 'Corolla']
model_filtered = df[df['Model'].isin(models_list)]
print("üöó –°—Ç—Ä–æ–∫–∏ —Å Model in ['Focus', 'Corolla']:")
print(model_filtered, "\n")

# 4 –ù–æ–º–µ—Ä–∞ –∏–∑ —Å–ø–∏—Å–∫–∞
numbers_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
carnumber_filtered = df.loc[df.index.isin(numbers_list)]
print("üöó –°—Ç—Ä–æ–∫–∏ —Å CarNumber in [Y7689C197RUS, ..., X758HY197RUS]:")
print(carnumber_filtered, "\n")
```

## make the aggregations with the make and the model

```python
# 1 –ú–µ–¥–∏–∞–Ω–Ω—ã–µ —à—Ç—Ä–∞—Ñ—ã –ø–æ –º–∞—Ä–∫–µ
median_by_make = df.groupby('Make')['Fines'].median()
print("üìä –ú–µ–¥–∏–∞–Ω–Ω—ã–µ —à—Ç—Ä–∞—Ñ—ã –ø–æ Make:")
print(median_by_make, "\n")

# 2 –ú–µ–¥–∏–∞–Ω–Ω—ã–µ —à—Ç—Ä–∞—Ñ—ã –ø–æ –º–∞—Ä–∫–µ –∏ –º–æ–¥–µ–ª–∏
median_by_make_model = df.groupby(['Make', 'Model'])['Fines'].median()
print("üìä –ú–µ–¥–∏–∞–Ω–Ω—ã–µ —à—Ç—Ä–∞—Ñ—ã –ø–æ Make –∏ Model:")
print(median_by_make_model, "\n")

# 3 –ö–æ–ª-–≤–æ —à—Ç—Ä–∞—Ñ–æ–≤ –ø–æ –º–∞—Ä–∫–µ –∏ –º–æ–¥–µ–ª–∏
count_by_make_model = df.groupby(['Make', 'Model'])['Fines'].count()
print("üìä –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —à—Ç—Ä–∞—Ñ–æ–≤ –ø–æ Make –∏ Model:")
print(count_by_make_model, "\n")

# 4 –ú–∏–Ω/–º–∞–∫—Å —à—Ç—Ä–∞—Ñ—ã
minmax_by_make_model = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
print("üìä –ú–∏–Ω/–ú–∞–∫—Å —à—Ç—Ä–∞—Ñ–æ–≤ –ø–æ Make –∏ Model:")
print(minmax_by_make_model, "\n")

# 5 –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —à—Ç—Ä–∞—Ñ–æ–≤
std_by_make_model = df.groupby(['Make', 'Model'])['Fines'].std()
print("üìä –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —à—Ç—Ä–∞—Ñ–æ–≤ –ø–æ Make –∏ Model:")
print(std_by_make_model, "\n")
```

## make the aggregations with the car number

```python
# 1 –°–∞–º—ã–µ —á–∞—Å—Ç—ã–µ –Ω–∞—Ä—É—à–∏—Ç–µ–ª–∏
violators_by_count = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
print("üö® –ú–∞—à–∏–Ω—ã –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —à—Ç—Ä–∞—Ñ–æ–≤ (—É–±—ã–≤–∞–Ω–∏–µ):")
print(violators_by_count, "\n")

# 2 –í—Å–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ —Ç–æ–ø-1 –Ω–∞—Ä—É—à–∏—Ç–µ–ª—é
top1_car_by_count = violators_by_count.index[0]
top1_car_rows = df.loc[top1_car_by_count]
print(f"üîç –í—Å–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ —Ç–æ–ø-1 –º–∞—à–∏–Ω–µ –ø–æ —á–∏—Å–ª—É —à—Ç—Ä–∞—Ñ–æ–≤ ({top1_car_by_count}):")
print(top1_car_rows, "\n")

# 3 –°–∞–º—ã–µ –¥–æ—Ä–æ–≥–∏–µ –Ω–∞—Ä—É—à–∏—Ç–µ–ª–∏
violators_by_sum = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
print("üí∞ –ú–∞—à–∏–Ω—ã –ø–æ —Å—É–º–º–µ —à—Ç—Ä–∞—Ñ–æ–≤ (—É–±—ã–≤–∞–Ω–∏–µ):")
print(violators_by_sum, "\n")

# 4 –í—Å–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ —Ç–æ–ø-1 –ø–æ —Å—É–º–º–µ
top1_car_by_sum = violators_by_sum.index[0]
top1_car_sum_rows = df.loc[top1_car_by_sum]
print(f"üîç –í—Å–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ —Ç–æ–ø-1 –º–∞—à–∏–Ω–µ –ø–æ —Å—É–º–º–µ —à—Ç—Ä–∞—Ñ–æ–≤ ({top1_car_by_sum}):")
print(top1_car_sum_rows, "\n")

# 5. –ü—Ä–æ–≤–µ—Ä–∫–∞: –µ—Å—Ç—å –ª–∏ –Ω–æ–º–µ—Ä–∞ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –º–æ–¥–µ–ª—è–º–∏
car_model_variation = df.reset_index().groupby('CarNumber')['Model'].nunique()
cars_with_multiple_models = car_model_variation[car_model_variation > 1]
# –í—ã–≤–æ–¥ (–ø–æ –∂–µ–ª–∞–Ω–∏—é, –º–æ–∂–Ω–æ —á–µ—Ä–µ–∑ print –∏–ª–∏ display –≤ Jupyter):
print("‚ùì –ú–∞—à–∏–Ω—ã, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ä–∞–∑–Ω—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏:")
print(cars_with_multiple_models)
```
